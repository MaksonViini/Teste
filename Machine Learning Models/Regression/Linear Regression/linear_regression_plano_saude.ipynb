```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/MaksonViini/Teste/master/Machine%20Learning%20Models/Datasets/plano_saude.csv')
```

```python
data.head()
```

```python
X = data.iloc[:, 0].values
y = data.iloc[:, 1].values
```

```python
corr = np.corrcoef(X, y)
corr
```

```python
data.corr()
```

```python
# Plotando a linearidade dos dados
plt.plot(X, y)
plt.grid()
plt.show()
```

```python
# Plotando a linearidade dos dados em modo scatter
plt.scatter(X, y)
plt.grid()
plt.show()
```

## Treinando o Modelo

```python
from sklearn.linear_model import LinearRegression
```

```python
X = X.reshape(-1, 1)
```

```python
lr = LinearRegression().fit(X, y)
```

### Regressao, y = ax + b

```python
# intercepto B
# Intercepto e onde comeca a reta
print('Intercepto: ', lr.intercept_)

# Coeficiente A
# Coefiente e quanto a reta varia de acordo com o valor de X, declive da linha
print('Coefiente: ', lr.coef_)
```

```python
lr.score(X.reshape(-1, 1), y.reshape(-1, 1))
```

## Visualizando a reta da regressao

```python
plt.scatter(X, y)
plt.plot(X, lr.predict(X))
plt.grid()
plt.show()
```

## Fazendo predicoes

```python
age = [np.array(np.random.randint(30, 50, 10)).reshape(-1, 1)]

for i in age:
    predict = lr.predict(i)
    print(f'A pessoa com idade de {i} tem um custo previsto do plano de R$ {predict.round(2)}')
```

## y = ax + b onde; y = _coef_ * x + _intercept_

```python
# y = ax + b
predict1 = lr.coef_ * 40 + lr.intercept_
print(predict1)
```

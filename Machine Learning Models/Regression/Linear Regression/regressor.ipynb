```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Plotar graficos no jupyter
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = np.linspace(0, 9, 10).reshape(-1, 1)
y = np.linspace(0, 27, 10).reshape(-1, 1)
```

```python
# Instanciando a Regressao Linear
lr = LinearRegression().fit(X, y)
```

```python
# Coeficiente e intercepto
print('Coeficiente: ', lr.coef_)
print('Intercept: ', lr.intercept_)
```

```python
# Grafico dos valores
plt.plot(X, y)
plt.grid(True)
plt.show()
```

## Regressao com base de dados do sklearn datasets

```python
# Criando nova base de dados
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=250, n_features=1, noise=21, random_state=42)

plt.scatter(X, y)
plt.show()
```

```python
lr = LinearRegression().fit(X, y)

# Coeficiente e intercepto
print('Coeficiente: ', lr.coef_)
print('Intercept: ', lr.intercept_)
```

```python
# Gerando valores para plotar o grafico
linha_gerada = np.linspace(-3, 3, 500).reshape(-1, 1)
```

```python
# Gerando a predicao do modelo
plt.plot(X, y, '^')
plt.plot(linha_gerada, lr.predict(linha_gerada))
plt.grid(True)
plt.show()
```

## Criando modelo com treino e teste

```python
# Separando entre valores de treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
lr_2 = LinearRegression().fit(X_train, y_train)
```

```python
# Coeficiente e intercepto
print('Coeficiente: ', lr.coef_)
print('Intercept: ', lr.intercept_)
```

```python
# Final
plt.plot(X_train, y_train, '^')
plt.plot(X_test, y_test, 'v')
plt.plot(linha_gerada, lr_2.predict(linha_gerada))
plt.grid()
plt.show()
```
